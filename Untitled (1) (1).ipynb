```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\retail_store_sales.csv")
y = df.select_dtypes(include='number').iloc[:50, 0]
plt.scatter(range(len(y)), y)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
customer_df = df.groupby('Customer ID').agg({
    'Price Per Unit':'sum',
    'Category': 'count',
    'Item': 'count'
}).reset_index()
customer_df.columns = ['Customer ID', 'price per unit', 'Category', 'Item']
customer_df.head()
```

```python
X = customer_df[['price per unit','Category','Item']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
customer_df['Class_Label'] = kmeans.fit_predict(X_scaled)

customer_df.head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    customer_df['Item'],
    customer_df['Category'],
    c=customer_df['Class_Label'],
    cmap='viridis'
)

plt.xlabel("Number of Items Purchased")
plt.ylabel("Number of Categories Purchased")
plt.title("Customer Segmentation using K-Means")
plt.colorbar(label='Class Label')

plt.show()
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\retail_store_sales.csv")
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
snapshot_date = df['Transaction Date'].max() + pd.Timedelta(days=1)
rfm_df = df.groupby('Customer ID').agg({
    'Transaction Date': lambda x: (snapshot_date - x.max()).days,
    'Transaction ID': 'count',                                    
    'Total Spent': 'sum'                                          
}).reset_index()
rfm_df.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
print(rfm_df.head())
```

```python
customer_df = df.groupby('Customer ID').agg({
    'Location': lambda x: x.mode()[0] if not x.mode().empty else 'Unknown',
    'Payment Method': lambda x: x.mode()[0] if not x.mode().empty else 'Unknown'
}).reset_index()
customer_df.rename(columns={'Customer ID': 'CustomerID'}, inplace=True)
final_df = customer_df.merge(rfm_df, on='CustomerID')
print(final_df.head())
```

```python
X_generalized = final_df[
    ['Location','Payment Method',
     'Recency', 'Frequency', 'Monetary']
]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
X_generalized_df = pd.DataFrame(X_generalized) 
categorical_cols = X_generalized_df.select_dtypes(include=['object']).columns
numerical_cols = X_generalized_df.select_dtypes(exclude=['object']).columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ])
X_generalized_scaled = preprocessor.fit_transform(X_generalized_df)
```

```python
kmeans_general = KMeans(n_clusters=3, random_state=42)
final_df['Generalized_Class_Label'] = kmeans_general.fit_predict(X_generalized_scaled)

final_df.head()
```

```python
final_df.groupby('Generalized_Class_Label').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean'
})
```

```python
cluster_names = {
    0: "Low Value / Inactive",
    1: "Potential Customers",
    2: "Loyal High-Value",
    3: "Regular Customers"
}

final_df['Cluster_Name'] = final_df['Generalized_Class_Label'].map(cluster_names)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))

plt.bar(final_df['Cluster_Name'],
        final_df['Frequency'])

plt.xlabel("Customer Segments")
plt.ylabel("Average Buying Frequency")
plt.title("Customer Segmentation Based on Buying Behavior")
plt.xticks(rotation=20)

plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_generalized_scaled,
                             final_df['Generalized_Class_Label'])

sil_score
```

```python
print(f"Silhouette Score for Generalized K-Means Clustering: {sil_score:.3f}")
```
